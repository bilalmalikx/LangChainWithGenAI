```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
embeddings_1024

text="this is my embeddings text"
embeddings_result=embeddings_1024.embed_query(text)
embeddings_result
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_openai import ChatOpenAI, OpenAIEmbeddings


loader = TextLoader("speech.txt")
docs = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_docs = text_splitter.split_documents(docs)
final_docs
```

```python
# vector embeddings and vectore store

from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_docs,embeddings_1024)
db
```

```python
query=("To such a task we can dedicate our lives and our fortunes")
retrived_result=db.similarity_search(query)
print(retrived_result)
```
